#### 1. Sketch the likelihood function for each class (ref. Figure 2.1). Calculate the likelihood value for each class if x=5.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
 
def normfun(x,mu,sigma):
    pdf = np.exp(-((x - mu)**2)/(2*sigma**2)) / (sigma * np.sqrt(2*np.pi))
    return pdf
 
mu = 2
sigma =2
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 50)
y_sig = normfun(x,mu,sigma)
plt.plot(x, y_sig, "r-", linewidth=2)
plt.vlines(mu, 0, np.exp(-(mu - mu) ** 2 /(2* sigma **2))/(math.sqrt(2*math.pi)*sigma), colors = "c", linestyles = "dashed")
mu = 0
sigma =1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 50)
y_sig = normfun(x,mu,sigma)
plt.plot(x, y_sig, "r-", linewidth=2)
plt.vlines(mu, 0, np.exp(-(mu - mu) ** 2 /(2* sigma **2))/(math.sqrt(2*math.pi)*sigma), colors = "c", linestyles = "dashed")
plt.xlabel('x')
plt.ylabel('Likelihood')
plt.grid(True)
plt.show()
```

```python
pdf51 = normfun(5,0,1)
pdf52 = normfun(5,2,2)
print ('While x=5, the likelihood value of class1 is', pdf51)
print ('While x=5, the likelihood value of class2 is', pdf52)
```

#### 2. Sketch the evidence distribution. Calculate the evidence value if x=5.

```python
def evidence(x):
    pdf1 = normfun(x,0,1)
    pdf2 = normfun(x,2,2)
    evidence = pdf1*0.6+pdf2*0.4
    return evidence

x = np.linspace(-5, 10, 50)
y = evidence(x)
plt.plot(x, y, "r-", linewidth=2)
plt.xlabel('x')
plt.ylabel('evidence')
plt.grid(True)
plt.show()
```

```python
evidence5 = evidence(5)
print ("While x=5, the evidence value is",evidence5)
```

#### 3. Sketch the posterior probability function for each class (ref. Figure 2.2). Calculate the posterior probability value for each class if x=5.

```python
def posterior1(x,pdf,evidence):
    posterior1 = (pdf1*0.6)/evidence
    return posterior1

def posterior2(x,pdf,evidence):
    posterior2 = pdf2*0.4/evidence
    return posterior2

x = np.linspace(-5, 10, 50)
pdf1 = normfun(x,0,1)
pdf2 = normfun(x,2,2)
evidence=evidence(x)
y1 = posterior1(x,pdf1,evidence)
y2 = posterior2(x,pdf2,evidence)
plt.plot(x, y1, "r-", linewidth=2)
plt.plot(x, y2, "r-", linewidth=2)
plt.xlabel('x')
plt.ylabel('Posterior')
plt.grid(True)
plt.show()
```

#### 4. Sketch the likelihood ratio function (ref. Figure 2.3). Calculate the likelihood ratio value if x=5.

```python
x = np.linspace(-5, 10, 50)
pdf1 = normfun(x,0,1)
pdf2 = normfun(x,2,2)
ratio = pdf1/pdf2
plt.plot(x, ratio, "r-", linewidth=2)
plt.xlabel('x')
plt.ylabel('Likelihood ratio')
plt.grid(True)
plt.show()

ratio5 = normfun(5,0,1)/normfun(5,2,2)
print('While x=5, the Likelihood ratio is', ratio5)
```
